Linear Discriminant Analysis (LDA) es una técnica para transformación de datos, cuyo objetivo es maximizar la separación entre 2 clases creando distribuciones gausianas. La documentación de este método se encuentra en https://scikit-learn.org/0.16/modules/generated/sklearn.lda.LDA.html

Para este ejercicio, probar LDA para reducir dimensiones del dataset y entrenar 2 modelos: SVC y XGBoost. ¿Que `accuracy` logra encontrar ahora y a cuantas dimensiones se redujo el problema?

```python
columnas_df2=np.unique(y)
```

```python
class_feature_means = pd.DataFrame(columns=columnas_df2)
for c, rows in df_2.groupby('Class'):
    class_feature_means[c] = rows.mean()
class_feature_means
```

```python
within_class_scatter_matrix = np.zeros((18,18))
for c, rows in df_2.groupby('Class'):
    rows = rows.drop(['Class'], axis=1)
    
    s = np.zeros((18,18))
for index, row in rows.iterrows():
        x, mc = row.values.reshape(18,1), class_feature_means[c].values.reshape(18,1)
        s += (x - mc).dot((x - mc).T)
        within_class_scatter_matrix += s
```

```python
feature_means = df_2.mean()
between_class_scatter_matrix = np.zeros((18,18))
for c in class_feature_means:    
    n = len(df_2.loc[df_2['Class'] == c].index)
    
    mc, m = class_feature_means[c].values.reshape(18,1), feature_means.values.reshape(18,1)
    
    between_class_scatter_matrix += n * (mc - m).dot((mc - m).T)
```

```python
eigen_values, eigen_vectors = np.linalg.eig(np.linalg.inv(within_class_scatter_matrix).dot(between_class_scatter_matrix))
```

```python
pairs = [(np.abs(eigen_values[i]), eigen_vectors[:,i]) for i in range(len(eigen_values))]
pairs = sorted(pairs, key=lambda x: x[0], reverse=True)
for pair in pairs:
    print(pair[0])
```

```python
eigen_value_sums = sum(eigen_values)
print('Explicando el valor de varianza')
for i, pair in enumerate(pairs):
    print('Eigenvector {}: {}'.format(i, (pair[0]/eigen_value_sums).real))
```

```python
w_matrix = np.hstack((pairs[0][1].reshape(18,1), pairs[1][1].reshape(18,1))).real
```

```python
X_lda = np.array(X.dot(w_matrix))
```

```python
le = LabelEncoder()
y = le.fit_transform(df_2['Class'])
```
